# **Tests in Time Series Analysis**

In time series analysis, various statistical tests are used to analyze data that is ordered by time. These tests help to understand the characteristics of the time series, such as stationarity, seasonality, trend, and autocorrelation. Below are some of the most commonly used tests in time series analysis:

### 1. **Augmented Dickey-Fuller (ADF) Test**
   - **Purpose**: Tests for **stationarity** of a time series.
   - **Null Hypothesis**: The time series has a unit root (i.e., it is non-stationary).
   - **Alternative Hypothesis**: The time series is stationary.
   - **Usage**: A significant p-value (usually < 0.05) indicates stationarity. A non-significant p-value suggests that the series is non-stationary and may require differencing.

### 2. **Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test**
   - **Purpose**: Tests for stationarity in a time series, but with a different approach than ADF.
   - **Null Hypothesis**: The time series is stationary.
   - **Alternative Hypothesis**: The time series is non-stationary.
   - **Usage**: A significant p-value suggests non-stationarity. This test is often used in combination with the ADF test for more robust results.

### 3. **Phillips-Perron (PP) Test**
   - **Purpose**: Tests for unit roots in a time series (similar to the ADF test) but adjusts for autocorrelation and heteroskedasticity in errors.
   - **Null Hypothesis**: The time series has a unit root.
   - **Alternative Hypothesis**: The time series does not have a unit root (i.e., it is stationary).
   - **Usage**: Like ADF, a significant p-value suggests stationarity, and a non-significant p-value indicates non-stationarity.

### 4. **Ljung-Box Test**
   - **Purpose**: Tests for autocorrelation at multiple lags in a time series.
   - **Null Hypothesis**: There is no significant autocorrelation at any lag.
   - **Alternative Hypothesis**: There is significant autocorrelation at one or more lags.
   - **Usage**: A significant p-value (usually < 0.05) indicates that there is significant autocorrelation, which can suggest a need for a model like ARIMA.

### 5. **Jarque-Bera Test**
   - **Purpose**: Tests for normality of residuals in time series data.
   - **Null Hypothesis**: The data follows a normal distribution.
   - **Alternative Hypothesis**: The data does not follow a normal distribution.
   - **Usage**: A significant p-value suggests that the residuals are not normally distributed, which might suggest the need for a different transformation or model.

### 6. **Granger Causality Test**
   - **Purpose**: Tests whether one time series can predict another.
   - **Null Hypothesis**: Time series X does not Granger cause time series Y.
   - **Alternative Hypothesis**: Time series X Granger causes time series Y.
   - **Usage**: A significant p-value indicates that past values of one series are useful in predicting another.

### 7. **Unit Root Tests (e.g., ADF, PP, KPSS)**
   - **Purpose**: Tests whether a time series has a unit root (i.e., is non-stationary).
   - **Null Hypothesis**: The time series has a unit root.
   - **Alternative Hypothesis**: The time series does not have a unit root and is stationary.
   - **Usage**: Often performed before applying ARIMA or other time series models to ensure that the series is stationary.

### 8. **Hurst Exponent Test**
   - **Purpose**: Assesses the long-term memory of a time series. It is used to identify whether the time series follows a random walk or exhibits persistent behavior.
   - **Null Hypothesis**: The time series has no long-term memory.
   - **Alternative Hypothesis**: The time series exhibits long-term memory (persistent or anti-persistent).
   - **Usage**: A Hurst exponent value of 0.5 suggests a random walk, while values greater than 0.5 suggest a persistent, trending series.

### 9. **Cochrane-Orcutt Procedure**
   - **Purpose**: An iterative procedure to correct for autocorrelation in the residuals of a regression model.
   - **Null Hypothesis**: No autocorrelation is present.
   - **Alternative Hypothesis**: Autocorrelation is present, and the model needs correction.

### 10. **Anderson-Darling Test**
   - **Purpose**: A test for normality, which is often used to check if residuals of time series models are normally distributed.
   - **Null Hypothesis**: The residuals are normally distributed.
   - **Alternative Hypothesis**: The residuals are not normally distributed.

### 11. **Variance Ratio Test**
   - **Purpose**: Tests whether a time series is a random walk (non-stationary) or if it has some trend.
   - **Null Hypothesis**: The time series is a random walk.
   - **Alternative Hypothesis**: The time series is not a random walk.
   - **Usage**: A significant result suggests that the series might be non-stationary.

### 12. **Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF)**
   - **Purpose**: Used to analyze the autocorrelations in the time series. ACF measures the correlation between the series and its lags, while PACF measures the correlation between the series and its lags, accounting for intermediate lags.
   - **Usage**: These functions help in identifying the appropriate parameters (p, q) for ARIMA models.

# **Augmented Dickey-Fuller (ADF) Test**

#### **Description**:
The Augmented Dickey-Fuller (ADF) test is a statistical test used to determine whether a time series is stationary or has a unit root, which implies non-stationarity. Stationarity is a key assumption in many time series forecasting models (like ARIMA), and the ADF test helps in assessing whether a time series is suitable for such models or if it needs differencing to make it stationary.

In the ADF test, a unit root is tested for, which, if present, indicates non-stationarity (i.e., the mean, variance, and autocorrelation structure of the time series are not constant over time).

#### **Hypothesis**:
- **Null Hypothesis (H₀)**: The time series has a unit root (i.e., the time series is non-stationary).
- **Alternative Hypothesis (H₁)**: The time series does not have a unit root (i.e., the time series is stationary).

#### **Usage**:
- The ADF test is used to check the stationarity of a time series before applying models that assume stationarity (e.g., ARIMA).
- If the p-value from the test is small (typically less than 0.05), we reject the null hypothesis and conclude that the series is stationary.
- If the p-value is large (greater than 0.05), we fail to reject the null hypothesis and conclude that the series is non-stationary and may require differencing.

#### **Numerical Example**:
Let's consider a simple time series: `y = [1, 2, 3, 4, 5, 6, 7]`. We want to test whether this series is stationary using the ADF test.

```python
import numpy as np
from statsmodels.tsa.stattools import adfuller

# Example time series
time_series = [1, 2, 3, 4, 5, 6, 7]

# Perform Augmented Dickey-Fuller test
result = adfuller(time_series)

# Display results
print("ADF Statistic: ", result[0])
print("P-Value: ", result[1])
print("Critical Values: ", result[4])

# Interpretation of the p-value
if result[1] < 0.05:
    print("Reject the null hypothesis: The series is stationary.")
else:
    print("Fail to reject the null hypothesis: The series is non-stationary.")
```

---

**Expected Output**:
- **ADF Statistic**: A value indicating how much the time series deviates from being stationary.
- **P-Value**: A value indicating the probability of observing the test statistic under the null hypothesis.

If the p-value is below 0.05, we reject the null hypothesis (indicating the series is stationary).

#### **Advantages**:
- **Test for Stationarity**: The ADF test is widely used for checking the stationarity of time series data, which is crucial for many time series models like ARIMA.
- **Widely Accepted**: The ADF test is one of the most widely used methods to test for unit roots in time series.
- **Consideration for Autocorrelation**: The test accounts for autocorrelation in the time series by including lagged differences.

#### **Disadvantages**:
- **Assumption of Linearity**: The ADF test assumes a linear relationship in the data. It may not perform well with non-linear time series data.
- **Sensitivity to Model Parameters**: The choice of lag length and trend terms in the ADF test can significantly affect the results.
- **False Positives in Small Samples**: The test can sometimes fail to reject the null hypothesis (i.e., incorrectly indicate non-stationarity) when sample sizes are small.
---

---

### **Factors for Choosing a Test (Related to ADF Test)**:
- **Type of Data**: If you're working with time series data, ADF is a suitable test for checking stationarity.
- **Number of Groups**: ADF test is applicable to a single time series, not multiple groups.
- **Sample Size**: The ADF test is more reliable with larger sample sizes. Small samples may lead to inaccurate conclusions.
- **Distribution**: The ADF test assumes the data follows an approximate normal distribution. If the data is highly skewed, alternative methods like the KPSS test might be more suitable.
- **Relationship Type**: ADF tests for a unit root in the time series, so it is concerned with checking whether the time series is autocorrelated or not.

# **Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF)**

#### **Description**:
- **Autocorrelation Function (ACF)**: The ACF measures the correlation between a time series and its lagged versions. It helps determine if there are repeating patterns or if values in the series are correlated with each other over different time lags.
  
  - The ACF values range from -1 to 1. A positive ACF indicates a positive correlation, and a negative ACF indicates a negative correlation. An ACF of 0 suggests no correlation.
  
  - The ACF is used in time series analysis to determine the appropriate model for time series forecasting, such as ARIMA, by identifying the lag values that are highly correlated.

- **Partial Autocorrelation Function (PACF)**: The PACF measures the correlation between a time series and its lagged version, but after removing the effects of intermediate lags. In other words, it helps to identify the direct correlation of a time series with its previous values without the influence of other lags.

  - PACF is particularly useful for identifying the order of the autoregressive (AR) term in an ARIMA model.

#### **Hypothesis**:
- **Null Hypothesis (H₀)**: There is no significant autocorrelation or partial autocorrelation in the data.
- **Alternative Hypothesis (H₁)**: There is significant autocorrelation or partial autocorrelation in the data.

#### **Usage**:
- **ACF** is primarily used to determine the moving average (MA) order of ARIMA models, while **PACF** is used to determine the autoregressive (AR) order.
- Both ACF and PACF are used to understand the dependence structure of the time series, especially when deciding how to model the data using ARIMA or other models.
- ACF is useful for identifying how well the time series data is correlated with its past lags, while PACF identifies the significant lags that directly influence the current value, without the intermediary effects.

#### **Advantages**:
- **ACF**:
  - Helps identify the structure of the moving average (MA) process in time series models.
  - Provides insight into the correlation at multiple lags, aiding in model identification.
  
- **PACF**:
  - Provides insight into the direct effect of previous values on the current value, helping in identifying the autoregressive (AR) order.

#### **Disadvantages**:
- **ACF**:
  - ACF does not remove the influence of intermediate lags, so it can be less informative when identifying direct lags that impact the current value.
  - ACF can be misleading when the series is not stationary.

- **PACF**:
  - PACF can be complex to interpret, especially when there are high-order lags, leading to potential misinterpretations.
  - PACF assumes the series is stationary, which may not hold in real-world data without proper transformations.

#### **Python Code**:
Here’s how you can calculate and visualize ACF and PACF for a given time series:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Example time series data
data = [10, 12, 14, 15, 18, 19, 22, 25, 30, 31, 32]
time_series = pd.Series(data)

# Plot ACF and PACF
plt.figure(figsize=(12, 6))

# ACF Plot
plt.subplot(121)
plot_acf(time_series, ax=plt.gca(), lags=3)
plt.title("Autocorrelation Function (ACF)")

# PACF Plot
plt.subplot(122)
plot_pacf(time_series, ax=plt.gca(), lags=3)
plt.title("Partial Autocorrelation Function (PACF)")

plt.tight_layout()
plt.show()
```

- **Number of Groups**: ACF and PACF are used on univariate time series data, so they are suitable for single group analysis.
- **Sample Size**: ACF and PACF are more reliable with larger sample sizes.
- **Distribution**: Both ACF and PACF assume the data is stationary. If the series is non-stationary, transformations (like differencing) are needed before applying ACF/PACF.
- **Relationship Type**: ACF is useful for understanding the correlation structure of the data at multiple lags, while PACF is better for understanding the direct influence of a given lag, without the influence of other lags.

# **Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test**

#### **Description**:
The Kwiatkowski-Phillips-Schmidt-Shin (KPSS) test is a statistical test used to assess the stationarity of a time series. Unlike the Augmented Dickey-Fuller (ADF) test, which tests for a unit root (i.e., non-stationarity), the KPSS test is designed to test the null hypothesis that a time series is **stationary** around a deterministic trend or is **trend stationary**.

The KPSS test is used in time series analysis when you want to test the stationarity of the series with respect to a deterministic trend.

- **Stationarity**: A time series is said to be stationary if its properties (mean, variance, and autocorrelation) do not change over time.
- **Trend Stationarity**: If a time series is trend stationary, it can be made stationary by removing a deterministic trend.

#### **Hypothesis**:
- **Null Hypothesis (H₀)**: The time series is stationary around a deterministic trend (i.e., the time series is trend stationary).
- **Alternative Hypothesis (H₁)**: The time series is not stationary (i.e., it has a unit root or contains a stochastic trend).

#### **Usage**:
- The KPSS test is commonly used in combination with the ADF test. While ADF tests for a unit root (non-stationarity), KPSS tests for stationarity. The two tests are often used together to confirm the stationarity or non-stationarity of a time series.
  
- If ADF fails to reject the null hypothesis (non-stationary) and KPSS fails to reject the null hypothesis (stationary), then the time series is considered to be trend stationary.
  
- If both tests reject the null hypothesis, the series is considered to be non-stationary.

#### **Mathematical Algorithm**:
The KPSS test statistic is calculated by examining the residuals of the time series after removing the deterministic trend. The steps in the KPSS test are as follows:

1. Detrend the time series by removing the deterministic trend.
2. Compute the residuals from the detrended series.
3. Calculate the partial sum of the residuals.
4. Compute the test statistic as the ratio of the partial sum to the variance of the residuals.

---
#### **Advantages**:
1. **Simple Interpretation**: The KPSS test is straightforward to use and interpret. If the p-value is small, it indicates that the time series is likely non-stationary.
2. **Trend Stationarity Testing**: The KPSS test is useful for confirming if a series is trend stationary, providing complementary insights to the ADF test.
3. **Good for Time Series with Trends**: If you expect your time series to have a trend component, the KPSS test is a good choice to verify its stationarity.

---

#### **Disadvantages**:
1. **Sensitivity to Lags**: The KPSS test may not perform well with small sample sizes or when the number of lags is improperly chosen.
2. **Non-Stationary Data**: If the series is not trend stationary (i.e., it has a unit root), the KPSS test will often fail to reject the null hypothesis, even though the series is non-stationary.
3. **Does Not Handle Stochastic Trends**: Unlike the ADF test, which can test for both trend and stochastic stationarity, the KPSS test only works for trend stationarity.
---

#### **Python Code**:

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import kpss
import matplotlib.pyplot as plt

# Example time series data
data = [10, 12, 14, 15, 18, 19, 22, 25, 30, 31, 32]
time_series = pd.Series(data)

# Perform KPSS Test
statistic, p_value, lags, critical_values = kpss(time_series, regression='c')

# Display results
print(f"KPSS Statistic: {statistic}")
print(f"P-Value: {p_value}")
print(f"Lags: {lags}")
print(f"Critical Values: {critical_values}")

# Interpretation
if p_value < 0.05:
    print("Reject null hypothesis - The series is not stationary.")
else:
    print("Fail to reject null hypothesis - The series is stationary.")
```

---

#### **Factors for Choosing KPSS Test**:
- **Type of Data**: KPSS is used for time series data that is suspected to be trend stationary.
- **Number of Groups**: This test is applied to a single time series (univariate data).
- **Sample Size**: Works best with moderate to large sample sizes. Small sample sizes can lead to unreliable results.
- **Distribution**: Works for stationary or trend stationary time series. The data should ideally be approximately normally distributed, though the test is somewhat robust to mild deviations.
- **Relationship Type**: Useful for testing if the time series has a deterministic trend, which is a common structure in many economic and financial time series.

# **Phillips-Perron (PP) Test**



#### **Description**:
The Phillips-Perron (PP) test is a statistical test used to check the presence of a unit root in a time series. It is an extension of the Augmented Dickey-Fuller (ADF) test, designed to provide a more robust approach to testing for unit roots, especially when there are autocorrelations or heteroskedasticity in the errors. The PP test corrects for serial correlation and time-varying volatility in the error term, which can often lead to unreliable test results.

The PP test is used to determine whether a time series is stationary or contains a unit root (i.e., it is non-stationary). If a time series contains a unit root, it indicates the series is non-stationary and needs to be transformed before modeling.

- **Unit Root**: If a time series contains a unit root, it means that the series is non-stationary, and its mean and variance are not constant over time.
- **Stationarity**: A time series is stationary if it has a constant mean and variance over time.

#### **Hypothesis**:
- **Null Hypothesis (H₀)**: The time series has a unit root (i.e., the series is non-stationary).
- **Alternative Hypothesis (H₁)**: The time series does not have a unit root (i.e., the series is stationary).

#### **Usage**:
The PP test is commonly used in econometrics and time series analysis to test for the presence of a unit root. It is particularly useful when dealing with time series data that might exhibit autocorrelation or heteroscedasticity. The PP test is generally used in the context of macroeconomics, finance, and environmental sciences to assess the stability of a series over time.

- **When to use**: The PP test is often preferred when the data has heteroscedasticity or autocorrelation issues. It provides more reliable results than the ADF test in such cases.

#### **Mathematical Algorithm**:
The PP test modifies the ADF test by adjusting the test statistic to account for serial correlation and heteroscedasticity. The PP test uses a non-parametric correction to the error term to control for autocorrelation and heteroscedasticity.

---

#### **Advantages**:
1. **Robustness**: The Phillips-Perron test corrects for autocorrelation and heteroscedasticity, making it more robust than the ADF test, especially when dealing with time series data that exhibit these issues.
2. **More Reliable for Econometric Data**: The PP test is particularly useful in econometric and financial data where heteroscedasticity or autocorrelation is often present.
3. **Non-Parametric**: The PP test is a non-parametric test, which makes fewer assumptions about the underlying data distribution, unlike the ADF test.

---

#### **Disadvantages**:
1. **Requires a Larger Sample Size**: The PP test requires a relatively larger sample size for reliable results compared to some other unit root tests.
2. **Less Common in Practice**: While the ADF test is more commonly used, the Phillips-Perron test is less widely adopted, making it harder to interpret in some contexts.
3. **Sensitivity to Model Specification**: The PP test can be sensitive to the choice of lag length and model specification, which can impact the results.

---

#### **Factors for Choosing the PP Test**:
- **Type of Data**: The PP test is suitable for univariate time series data. It is particularly used when dealing with economic, financial, or environmental time series that exhibit autocorrelation or heteroscedasticity.
- **Number of Groups**: The PP test is for a single time series, and it tests for a unit root in that series.
- **Sample Size**: The test performs better with larger sample sizes. Small sample sizes may lead to unreliable results.
- **Distribution**: The test is robust to heteroscedasticity and autocorrelation, making it useful for time series data with these characteristics.
- **Relationship Type**: The PP test is used when there is a need to check for unit roots (non-stationarity) in time series data. It is often used in combination with other tests like the ADF test.

# **Ljung-Box Test**

#### **Description**:
The Ljung-Box test is a statistical test used to check whether any of a group of autocorrelations of a time series are significantly different from zero. It is specifically designed to test for the presence of autocorrelation in a time series data, which is a common assumption when building time series models. The Ljung-Box test is often used to determine if the residuals of a model are random, as part of the model diagnostics process.

The Ljung-Box test is an extension of the Box-Pierce test and is more powerful for small sample sizes. It is a portmanteau test that checks whether a specified number of autocorrelations (lags) of the time series are zero.

- **Autocorrelation**: The correlation of a time series with its own past values.
- **White Noise**: A time series with no autocorrelation (i.e., a random sequence of values).

```python
import numpy as np
import pandas as pd
from statsmodels.stats.diagnostic import acorr_ljungbox

# Generate sample time series data
np.random.seed(42)
data = np.random.randn(100)

# Perform Ljung-Box test on the data
lb_stat, p_value = acorr_ljungbox(data, lags=10)



# Display results
print(f"Ljung-Box Statistic: {lb_stat}")
print(f"P-Value: {p_value}")

Interpretation
if p_value < 0.05:
    print("Reject null hypothesis - Significant autocorrelation present.")
else:
    print("Fail to reject null hypothesis - No significant autocorrelation.")
```

---

#### **Advantages**:
1. **Model Diagnostics**: It helps in diagnosing whether the residuals from a time series model (e.g., ARIMA) are autocorrelated or not.
2. **Robustness**: The Ljung-Box test is more powerful than the Box-Pierce test for small sample sizes, making it more reliable in practice.
3. **White Noise Verification**: It can be used to verify if a time series data is white noise, which is important when validating assumptions about randomness.

---

#### **Disadvantages**:
1. **Assumes Stationarity**: The test assumes that the time series is stationary, so it may not be appropriate for non-stationary time series without transformation.
2. **Sensitivity to Lags**: The test statistic's power can be influenced by the number of lags chosen. Too few lags might not capture the autocorrelation, while too many lags may lead to overfitting.
3. **False Positives**: Like other statistical tests, the Ljung-Box test can result in false positives if the sample size is small or if the data contains trends or non-stationarity.

---

#### **Factors for Choosing the Ljung-Box Test**:
- **Type of Data**: The test is suitable for stationary time series. Non-stationary series may need to be transformed first.
- **Number of Groups**: This is a univariate test, typically applied to a single time series.
- **Sample Size**: The Ljung-Box test can be sensitive to small sample sizes, so larger sample sizes provide more reliable results.
- **Distribution**: The test assumes that the series is stationary and that residuals follow a white noise process.
- **Relationship Type**: The Ljung-Box test is used when you suspect autocorrelation in the time series or residuals from a model. It is especially useful in model diagnostics.

# **Hurst Exponent Test**

---

The Hurst Exponent (H) is a measure of the long-term memory of a time series. It helps identify the type of behavior exhibited by the time series data: whether it exhibits persistence (trending), mean-reverting, or random behavior. The Hurst exponent is commonly used in the context of financial time series, hydrology, and geophysics to understand the tendency of time series to persist or revert.

---

### **Description:**
The Hurst exponent (H) is calculated to determine the "memory" of a time series:
- **If H > 0.5**: The time series exhibits a **persistent behavior** (trending). It means that an increase is likely to be followed by an increase (a trend).
- **If H = 0.5**: The time series behaves like a **random walk** (i.e., Brownian motion).
- **If H < 0.5**: The time series exhibits **mean-reverting behavior**. It means that an increase is likely to be followed by a decrease.

The Hurst exponent test is based on the rescaled range (R/S) analysis, which computes the ratio of the range to the standard deviation over a time period.

---

### **Hypothesis:**

- **Null Hypothesis (H₀)**: The time series follows a random walk, i.e., H = 0.5 (no persistence or mean-reversion).
- **Alternative Hypothesis (H₁)**: The time series does not follow a random walk (H ≠ 0.5), and it may exhibit persistence or mean-reversion.

---

### **Usage:**
- **Financial Market Analysis**: In finance, the Hurst exponent is used to identify the predictability of stock prices, returns, and other financial data.
- **Environmental Science**: It is used in hydrology to assess the behavior of streamflow or rainfall data.
- **Economics**: Used to model trends in economic data, inflation, GDP growth, etc.
