# **Central Tendency Measures (Mean, Median and Mode)**

image.png

image.png

```python
data = [4, 8, 6, 5, 3, 7]
mean = sum(data) / len(data)
print("Mean:", mean)
```

image.png

image.png

```python
data = [4, 8, 6, 5, 3, 7]
sorted_data = sorted(data)
n = len(sorted_data)

if n % 2 == 0:  # Even number of data points
    median = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
else:  # Odd number of data points
    median = sorted_data[n // 2]

print("Median:", median)
```

image.png

```python
from collections import Counter

data = [4, 8, 6, 5, 4, 7]
frequency = Counter(data)
mode = [key for key, value in frequency.items() if value == max(frequency.values())]

print("Mode:", mode)
```

image.png


### **Comparison of Mean, Median, and Mode**
| Metric  | **Mean**          | **Median**        | **Mode**             |
|---------|--------------------|-------------------|----------------------|
| **Sensitivity** | Affected by outliers. | Robust to outliers.  | No sensitivity to outliers. |
| **Applicability**| Numerical data.       | Skewed numerical data.| Categorical or frequent values. |
| **Ease of Use**  | Easy to calculate.    | Slightly harder due to sorting. | Simple for small datasets.  |


# **Variance and Standard Deviation, Skewness, and Kurtosis**

image.png

image.png

```python
import numpy as np

data = [4, 8, 6, 5, 3, 7]
variance = np.var(data)
std_dev = np.std(data)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
from scipy.stats import skew

data = [4, 8, 6, 5, 3, 7]
skewness = skew(data)

print("Skewness:", skewness)
```

image.png

image.png

- **High Kurtosis (Leptokurtic):** Heavy tails; more extreme outliers.

- **Low Kurtosis (Platykurtic):** Light tails; fewer extreme outliers.

- **Normal Kurtosis (Mesokurtic):** Kurtosis ≈ 3 (standard normal distribution).

image.png

```python
from scipy.stats import kurtosis

data = [4, 8, 6, 5, 3, 7]
kurt = kurtosis(data, fisher=True)  # Fisher=True adjusts to subtract 3

print("Kurtosis:", kurt)
```

image.png

### **Summary of Metrics**

| Metric                | **Variance**             | **Standard Deviation** | **Skewness**               | **Kurtosis**              |
|------------------------|--------------------------|-------------------------|----------------------------|---------------------------|
| **Purpose**            | Measure of spread.       | Measure of spread.      | Symmetry of distribution.  | Tailedness of distribution. |
| **Advantages**         | Precise measure.         | Intuitive to interpret. | Detects asymmetry.         | Identifies heavy/light tails. |
| **Disadvantages**      | Squared units are less intuitive. | Sensitive to outliers.  | May not show degree of asymmetry. | Complex to interpret.       |
