### **Non-Parametric Tests**

Non-parametric tests are statistical tests that do not assume a specific distribution for the data, making them useful when the data does not meet the assumptions required for parametric tests (e.g., normality, homogeneity of variances). These tests are particularly useful when dealing with ordinal data, nominal data, or when the sample size is small.

---

### **1. Common Non-Parametric Tests:**

#### **a. Mann-Whitney U Test** (for two independent groups)
- **Description**: The Mann-Whitney U test is used to determine if there is a difference between two independent groups. It is the non-parametric counterpart of the independent two-sample t-test.
- **Null Hypothesis**: The distributions of both groups are equal.
- **Alternative Hypothesis**: The distributions of both groups are not equal.

#### **b. Wilcoxon Signed-Rank Test** (for paired data)
- **Description**: The Wilcoxon signed-rank test is used to compare two related samples. It is the non-parametric equivalent of the paired t-test.
- **Null Hypothesis**: The median differences between paired observations are zero.
- **Alternative Hypothesis**: The median differences between paired observations are not zero.

#### **c. Kruskal-Wallis H Test** (for more than two independent groups)
- **Description**: The Kruskal-Wallis H test is an extension of the Mann-Whitney U test for comparing more than two independent groups. It is the non-parametric equivalent of one-way ANOVA.
- **Null Hypothesis**: The populations from which the samples are drawn have the same distribution.
- **Alternative Hypothesis**: At least one of the populations has a different distribution.

#### **d. Friedman Test** (for more than two related groups)
- **Description**: The Friedman test is a non-parametric alternative to the repeated measures ANOVA. It is used to test for differences in treatments across multiple related groups.
- **Null Hypothesis**: The distributions of all related groups are identical.
- **Alternative Hypothesis**: The distributions of at least one group differ from the others.

#### **e. Chi-Square Test**
- **Description**: The Chi-Square test is used to test the relationship between categorical variables. It can be used for testing the goodness of fit or testing the independence between variables.
- **Null Hypothesis**: There is no association between the variables.
- **Alternative Hypothesis**: There is an association between the variables.

#### **f. Kolmogorov-Smirnov Test**
- **Description**: The Kolmogorov-Smirnov test is used to compare a sample with a reference probability distribution or to compare two samples. It assesses the largest difference between the empirical cumulative distribution functions of the sample and the reference distribution.
- **Null Hypothesis**: The sample follows the reference distribution (or both samples are drawn from the same distribution).
- **Alternative Hypothesis**: The sample does not follow the reference distribution (or the samples are from different distributions).

#### **g. Spearman’s Rank Correlation**
- **Description**: The Spearman rank correlation test is used to measure the strength and direction of association between two ranked variables.
- **Null Hypothesis**: There is no correlation between the two variables.
- **Alternative Hypothesis**: There is a correlation between the two variables.

---

### **2. Advantages of Non-Parametric Tests:**

- **No Assumption of Normality**: Non-parametric tests do not require the assumption that the data follows a normal distribution, making them more flexible.
- **Robust to Outliers**: Non-parametric tests are less sensitive to outliers, as they often rely on ranks or medians rather than means.
- **Works for Ordinal Data**: These tests can be used for ordinal data or data that doesn't fit the assumptions of parametric tests.
- **Small Sample Sizes**: Non-parametric tests can be used with small sample sizes where parametric tests may not be valid due to lack of sufficient data.

---

### **3. Disadvantages of Non-Parametric Tests:**

- **Less Powerful**: Non-parametric tests are generally less powerful than parametric tests if the assumptions of parametric tests are met. This means that they are less likely to detect a true effect when one exists.
- **Less Information**: Non-parametric tests do not provide estimates of the parameters (such as means and variances) of the population, which may be important for interpretation.
- **Rank-Based**: Some non-parametric tests, like the Mann-Whitney U test, use ranks rather than actual data values, which can sometimes reduce the precision of the test.
- **Limited to Specific Data Types**: Non-parametric tests are mostly suitable for ordinal or non-normally distributed interval data. They are not always applicable to all types of data.

---

### **4. When to Use Non-Parametric Tests:**

- When the data is **not normally distributed**, and you cannot transform it into a normal distribution.
- When you have **ordinal data** or **categorical data**.
- When you have **small sample sizes** where the central limit theorem does not apply.
- When you are comparing **medians** or **ranks** rather than means.

### **5. Python Implementation Examples**:

---


```python

#### **Mann-Whitney U Test**:

import numpy as np
from scipy.stats import mannwhitneyu

group1 = np.array([85, 88, 90, 91, 87])
group2 = np.array([75, 78, 72, 80, 79])

stat, p_value = mannwhitneyu(group1, group2)
print(f"Mann-Whitney U Test Statistic: {stat}")
print(f"P-Value: {p_value}")
print("                                           ")

#### **Wilcoxon Signed-Rank Test**:

from scipy.stats import wilcoxon

before = np.array([60, 65, 70, 75, 80])
after = np.array([65, 70, 75, 80, 85])

stat, p_value = wilcoxon(before, after)
print(f"Wilcoxon Signed-Rank Test Statistic: {stat}")
print(f"P-Value: {p_value}")
print("                                           ")

#### **Kruskal-Wallis H Test**:

from scipy.stats import kruskal

group1 = np.array([85, 88, 90, 91, 87])
group2 = np.array([75, 78, 72, 80, 79])
group3 = np.array([65, 68, 70, 72, 74])

stat, p_value = kruskal(group1, group2, group3)
print(f"Kruskal-Wallis H Test Statistic: {stat}")
print(f"P-Value: {p_value}")
print("                                           ")

#### **Friedman Test**:

from scipy.stats import friedmanchisquare

group1 = np.array([80, 85, 90])
group2 = np.array([85, 90, 95])
group3 = np.array([75, 80, 85])

stat, p_value = friedmanchisquare(group1, group2, group3)
print(f"Friedman Test Statistic: {stat}")
print(f"P-Value: {p_value}")
print("                                           ")

#### **Chi-Square Test**:

from scipy.stats import chi2_contingency

# Contingency Table: Rows are different groups, columns are different categories.
data = np.array([[10, 20, 30], [6, 9, 17]])

stat, p_value, dof, expected = chi2_contingency(data)
print(f"Chi-Square Statistic: {stat}")
print(f"P-Value: {p_value}")
print("                                           ")

#### **Spearman Rank Correlation**:

from scipy.stats import spearmanr

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 3, 2, 1])

stat, p_value = spearmanr(x, y)
print(f"Spearman Rank Correlation: {stat}")
print(f"P-Value: {p_value}")
print("                                           ")
```

# **Mann-Whitney U Test**



---

### **1. Description**

The Mann-Whitney U Test (also known as the Wilcoxon rank-sum test) is a **non-parametric test** used to compare two independent groups to determine whether their population distributions differ. Unlike the t-test, it does not assume normality of data and works well with ordinal data or when data is not normally distributed. It is based on ranking the data rather than the raw data itself.

#### **Null Hypothesis (H₀):**
- The distributions of the two groups are identical.

#### **Alternative Hypothesis (H₁):**
- The distributions of the two groups are not identical.

This test is particularly useful when the sample sizes are small or when the data contains outliers.

# **Wilcoxon Signed-Rank Test**

---

### **1. Description**

The Wilcoxon Signed-Rank Test is a **non-parametric statistical test** used to compare two related samples, matched samples, or repeated measurements to assess whether their population mean ranks differ. It is used as an alternative to the paired t-test when the data does not satisfy the normality assumption. Instead of comparing the means, it compares the median differences.

#### **Null Hypothesis (H₀):**
- The median difference between the paired samples is zero.  
  H_0: The two distributions are identical.

#### **Alternative Hypothesis (H₁):**
- The median difference between the paired samples is not zero.  
  H_1: The two distributions are not identical.

This test is suitable for ordinal data or continuous data that violates normality assumptions.

#### **Types:**
1. **Two-Tailed Test:** Checks if there is any difference between the two distributions.
2. **One-Tailed Test:** Tests whether one distribution tends to have larger (or smaller) values than the other.

# **Kruskal-Wallis H Test**

---

### **1. Description**

The **Kruskal-Wallis H Test** is a non-parametric statistical test used to compare the medians of three or more independent groups to determine if they come from the same distribution. It is an extension of the Mann-Whitney U test to more than two groups and does not assume that the data is normally distributed. Instead, it works with ranked data.

#### **Null Hypothesis (H₀):**
- The populations from which the groups are drawn have the same median.  
  \( H_0: \text{The distributions are identical.} \)

#### **Alternative Hypothesis (H₁):**
- At least one of the group medians is different.  
  \( H_1: \text{At least one distribution differs from the rest.} \)

#### **Types of Kruskal-Wallis Test:**
- There is only one type, but the analysis is extended to post-hoc tests if the null hypothesis is rejected to determine which groups differ.

---

4. **Robust to Outliers**: Less sensitive to extreme values.

# **Friedman Test (for More Than Two Related Groups)**

---

### **1. Description**

The **Friedman Test** is a non-parametric test used to detect differences in treatments across multiple test attempts. It is used when there are three or more related groups (e.g., repeated measures or matched samples) and serves as an alternative to the repeated-measures ANOVA when the assumption of normality is not met.

#### **Null Hypothesis (H₀):**
- The distributions of all treatments (or groups) are identical.  
   H_0:The median ranks of the groups are equal.

#### **Alternative Hypothesis (H₁):**
- At least one treatment (or group) has a different distribution.  
  H_1: At least one group differs in median ranks.

The test compares the average ranks of groups and assesses whether differences are statistically significant.

---

### **4. Advantages**

1. **Non-Parametric Nature:** Suitable for non-normal data distributions.
2. **Handles Ordinal Data:** Works with ranked or ordinal data.
3. **More Than Two Groups:** Ideal for analyzing related samples across multiple treatments.
4. **Robust to Outliers:** Less sensitive to extreme values.

---

### **5. Disadvantages**

1. **Loss of Information:** Converts data to ranks, losing actual values.
2. **Assumes Similar Distributions:** Assumes similar shapes under the null hypothesis.

```python
import numpy as np
from scipy.stats import friedmanchisquare

# Data: Three groups (treatments) across four subjects
group_A = np.array([5, 6, 8, 7])
group_B = np.array([7, 6, 7, 5])
group_C = np.array([8, 9, 10, 9])

# Perform Friedman Test
friedman_stat, p_value = friedmanchisquare(group_A, group_B, group_C)

print("Friedman Test:")
print(f"Test Statistic: {friedman_stat:.2f}")
print(f"P-Value: {p_value:.4f}")
```

---

### **7. Types of Friedman Test**

The Friedman Test itself does not have subtypes, but it is often followed by **post-hoc tests** to identify specific group differences if the null hypothesis is rejected. Common post-hoc tests include:
1. **Nemenyi Test**: For pairwise comparisons between all groups.
2. **Wilcoxon Signed-Rank Test**: For pairwise comparisons between two specific groups.

---

### **Factors for Choosing a Test**

2. **Number of Groups:** Three or more related groups (e.g., repeated measures).
3. **Sample Size:** Effective with small to medium-sized datasets.
4. **Distribution:** Non-parametric; no normality assumption.
5. **Relationship Type:** Repeated measurements or matched groups.

# **Kolmogorov-Smirnov Test**

---

### **1. Description**

The **Kolmogorov-Smirnov (KS) Test** is a non-parametric test used to compare a sample with a reference probability distribution (one-sample KS test) or to compare two samples (two-sample KS test). It measures the maximum difference between the empirical distribution function (EDF) of the sample(s) and the cumulative distribution function (CDF) of the reference distribution or another sample.

#### **Null Hypothesis (H₀):**
- **One-Sample KS Test:** The sample data follows the specified reference distribution.
- **Two-Sample KS Test:** The two samples are drawn from the same distribution.

#### **Alternative Hypothesis (H₁):**
- **One-Sample KS Test:** The sample data does not follow the specified reference distribution.
- **Two-Sample KS Test:** The two samples are drawn from different distributions.

### **4. Advantages**

4. **Handles Small Samples:** Works effectively with small datasets.

---

### **5. Disadvantages**

1. **Only Detects Maximum Difference:** May miss small but significant differences across distributions.
2. **Requires Continuous Data:** Performance may degrade with discrete data.

### **6. Python Implementation**

#### **One-Sample KS Test**

```python
from scipy.stats import kstest, norm

# Sample data
data = [1.2, 1.8, 2.0, 2.4, 3.0]

# Perform One-Sample KS Test against a normal distribution
d_stat, p_value = kstest(data, 'norm', args=(2, 1))  # mean=2, std=1

print("One-Sample KS Test:")
print(f"D-Statistic: {d_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
```

#### **Two-Sample KS Test**

```python
from scipy.stats import ks_2samp

# Sample data
sample1 = [1.0, 1.5, 2.0, 2.5]
sample2 = [1.1, 1.6, 2.1, 2.6]

# Perform Two-Sample KS Test
d_stat, p_value = ks_2samp(sample1, sample2)

print("Two-Sample KS Test:")
print(f"D-Statistic: {d_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
```

### **7. Types of KS Test**
