# PARAMETRIC TESTS

**Parametric hypothesis tests** are statistical tests that make certain assumptions about the parameters (such as mean and variance) of the population distribution from which the data is sampled. These tests assume that the data follows a known distribution, usually the **normal distribution**, and they are based on sample data used to infer population parameters.

In general, **parametric tests** are more powerful than non-parametric tests because they use specific assumptions about the population distribution, leading to more accurate inferences when those assumptions are met.

### Key Features of Parametric Tests:
1. **Assumption of Normality**: Most parametric tests assume that the data comes from a normal distribution (or approximately normal distribution) in the population.
2. **Known or Assumed Parameters**: The tests often require knowledge or assumptions about population parameters, such as the population mean or standard deviation.
3. **Dependence on Sample Size**: For smaller sample sizes, parametric tests may not perform well, and in those cases, non-parametric tests (which don't rely on distribution assumptions) may be more appropriate.

---

### **Common Parametric Hypothesis Tests:**

1. **One-Sample t-Test**:
   - **Purpose**: Tests whether the mean of a sample is significantly different from a known population mean.
   - **Assumptions**: Data is approximately normally distributed, and the sample size is not extremely small.
   - **Null Hypothesis** (\( H_0 \)): The sample mean is equal to the population mean.
   - **Alternative Hypothesis** (\( H_1 \)): The sample mean is not equal to the population mean.

2. **Two-Sample t-Test**:
   - **Purpose**: Compares the means of two independent groups to see if they are significantly different from each other.
   - **Assumptions**: Data from both groups should be approximately normally distributed, and the two groups should have similar variances (for the pooled t-test).
   - **Null Hypothesis** (\( H_0 \)): The means of the two groups are equal.
   - **Alternative Hypothesis** (\( H_1 \)): The means of the two groups are different.

3. **Paired t-Test**:
   - **Purpose**: Compares the means of two related groups (e.g., measurements before and after an intervention) to determine if there is a significant difference.
   - **Assumptions**: Data from both groups should be approximately normally distributed, and the two samples are related.
   - **Null Hypothesis** (\( H_0 \)): There is no difference in the means of the two related groups.
   - **Alternative Hypothesis** (\( H_1 \)): There is a difference in the means of the two related groups.

4. **Z-Test (for Population Proportions or Means)**:
   - **Purpose**: Compares a sample mean (or proportion) with a population mean (or proportion) when the sample size is large, and the population standard deviation is known.
   - **Assumptions**: The sample comes from a normal distribution or the sample size is large enough for the Central Limit Theorem to apply.
   - **Null Hypothesis** (\( H_0 \)): The sample mean is equal to the population mean (or the sample proportion is equal to the population proportion).
   - **Alternative Hypothesis** (\( H_1 \)): The sample mean is not equal to the population mean (or the sample proportion is different from the population proportion).

5. **Analysis of Variance (ANOVA)**:
   - **Purpose**: Tests whether there is a significant difference between the means of three or more independent groups.
   - **Assumptions**: The groups are independent, data within each group is normally distributed, and the variances across the groups are equal.
   - **Null Hypothesis** (\( H_0 \)): All group means are equal.
   - **Alternative Hypothesis** (\( H_1 \)): At least one group mean is different.

6. **Regression Analysis (Linear Regression)**:
   - **Purpose**: Assesses the relationship between a dependent variable and one (or more) independent variable(s).
   - **Assumptions**: The relationship between the variables is linear, and the residuals (errors) are normally distributed with constant variance.
   - **Null Hypothesis** (\( H_0 \)): There is no linear relationship between the dependent and independent variables (i.e., the slope of the regression line is zero).
   - **Alternative Hypothesis** (\( H_1 \)): There is a significant linear relationship between the dependent and independent variables (i.e., the slope is non-zero).

---

### **Advantages of Parametric Hypothesis Tests:**
1. **Powerful**: When the assumptions are met, parametric tests are more powerful (i.e., they have a higher probability of correctly rejecting the null hypothesis) compared to non-parametric tests.
2. **Precise Estimates**: Parametric tests provide more precise estimates of population parameters, leading to more accurate conclusions.
3. **Widely Used**: Parametric tests are well-established and widely used in many scientific and social science fields.

---

### **Disadvantages of Parametric Hypothesis Tests:**
1. **Assumptions**: The primary disadvantage is that parametric tests depend on certain assumptions about the data (e.g., normality, known population parameters). If these assumptions are violated, the results may be misleading.
2. **Sensitive to Outliers**: Parametric tests can be sensitive to outliers or skewed data, which can distort the results.
3. **Not Suitable for Small Samples**: Parametric tests may not work well with small sample sizes unless the data is already normally distributed.

---

### **Conclusion:**
Parametric hypothesis tests are highly effective tools when the data meets certain criteria, such as normality and known population parameters. They are widely used in statistical analysis and provide powerful methods for hypothesis testing. However, they require that the assumptions be satisfied, and when the assumptions are violated, non-parametric tests (which do not require such strict assumptions) may be more appropriate.



### **T-Distribution**

#### **1. Description (In Detail):**  
The **T-Distribution**, also known as the **Student's T-Distribution**, is a probability distribution used in statistics when estimating population parameters (like the mean) from a small sample size, and the population standard deviation is unknown.  

It is similar to the normal distribution but has heavier tails, meaning it is more prone to producing values far from its mean. The shape of the T-Distribution depends on the **degrees of freedom (df)**, which is typically related to the sample size.  

- **Key Features of T-Distribution:**  
  1. **Symmetry:** It is symmetric around zero, just like the normal distribution.  
  2. **Degrees of Freedom (df):** As the degrees of freedom increase, the T-Distribution approaches the normal distribution.  
  3. **Use Case:** Commonly used in **T-Tests** for hypothesis testing and constructing confidence intervals.

### **T-Test**

#### **1. Description (In Detail):**  
A **T-Test** is a statistical test used to compare the means of one or two groups to determine if there is a statistically significant difference between them. It is based on the **t-distribution**, which is appropriate when the sample size is small and the population standard deviation is unknown.

The T-Test is commonly used in hypothesis testing to evaluate:
- Whether the sample mean is significantly different from a known value (one-sample t-test).
- Whether the means of two independent samples are significantly different (independent two-sample t-test).
- Whether the means of two related samples (before-and-after measurements) are significantly different (paired t-test).

The **T-Distribution** has heavier tails compared to the normal distribution, making it more sensitive to small sample sizes.

#### **4. Advantages :**

1. **Effective for Small Sample Sizes:**
   - The t-test is ideal for small sample sizes (less than 30) where the normal distribution assumption might not hold. The t-distribution accounts for the extra variability in small samples.
  
2. **Versatility:**
   - It can be used for various situations, including comparing a sample to a population, comparing two independent groups, and comparing two related groups (e.g., before-and-after scenarios).

3. **Robust to Minor Violations of Normality:**
   - The t-test is somewhat robust to slight violations of normality, especially for larger sample sizes.

4. **Widely Used and Understood:**
   - The t-test is one of the most commonly used statistical tests, making its results easy to interpret and explain.

---

#### **5. Disadvantages :**

1. **Assumes Normality:**
   - The t-test assumes that the data is normally distributed, especially for small sample sizes. If the data is not normally distributed, the results may be inaccurate. This assumption can be relaxed for larger sample sizes due to the Central Limit Theorem.

2. **Sensitive to Outliers:**
   - Like many parametric tests, the t-test is sensitive to outliers and extreme values. Outliers can heavily influence the mean and standard deviation, leading to misleading results.

3. **Requires Equal Variances (for Independent T-Test):**
   - The independent two-sample t-test assumes that both groups have equal variances (homoscedasticity). If the variances are significantly different (heteroscedasticity), the test may not be valid.

4. **Limited to Mean Comparisons:**
   - The t-test compares only the means of groups and does not account for other differences in data distributions (such as variance, skewness, etc.).

---

**6. Python Implementation:**
Here is a Python implementation for a One-Sample T-Test and Independent Two-Sample T-Test:

```python
import numpy as np
from scipy import stats

# One-Sample T-Test
data = np.array([78, 82, 80, 75, 76, 80, 83, 79, 81, 77])
population_mean = 75
t_statistic, p_value = stats.ttest_1samp(data, population_mean)

print("One-Sample T-Test:")
print(f"T-Statistic: {t_statistic:.2f}")
print(f"P-Value: {p_value:.4f}")

# Independent Two-Sample T-Test
group1 = np.array([65, 70, 75, 80, 85])
group2 = np.array([95, 100, 105, 110, 115])

t_statistic_indep, p_value_indep = stats.ttest_ind(group1, group2)

print("\nIndependent Two-Sample T-Test:")
print(f"T-Statistic: {t_statistic_indep:.2f}")
print(f"P-Value: {p_value_indep:.4f}")


# Paired T-Test
before_scores = np.array([60, 65, 70, 75, 80])
after_scores = np.array([65, 79, 80, 84, 81])

# Perform Paired T-Test
t_stat_paired, p_value_paired = stats.ttest_rel(before_scores, after_scores)
print("\nPaired T-Test:")
print(f"T-Statistic: {t_stat_paired:.2f}")
print(f"P-Value: {p_value_paired:.4f}")
```



#### **7. Types of T-Tests :**

1. **One-Sample T-Test:**
   - **Purpose:** Compares the mean of a sample to a known value (usually the population mean).
   - **Example Use Case:** Testing whether the average score of students in a class is different from a target score.
   - **Assumptions:** Data should be approximately normally distributed.

2. **Independent Two-Sample T-Test:**
   - **Purpose:** Compares the means of two independent groups to see if they are significantly different from each other.
   - **Example Use Case:** Comparing the test scores of two different classes.
   - **Assumptions:** The two groups should be independent and should ideally have equal variances (homogeneity of variance).

3. **Paired T-Test:**
   - **Purpose:** Compares the means of two related groups, often measuring the same subjects before and after a treatment or condition.
   - **Example Use Case:** Measuring blood pressure before and after a medication treatment in the same group of individuals.
   - **Assumptions:** The differences between paired observations should be approximately normally distributed.

---

### **Factors for Choosing a Test:**

|----------------------|-----------------------------------------------------------------------------|

# **Z-Test**

The Z-test is a statistical test used to determine whether there is a significant difference between sample data and a population or between two sample means. It assumes that the data follows a normal distribution and that the sample size is sufficiently large (typically n>=30)

### **4. Advantages:**

- **Efficient for Large Sample Sizes**: The Z-test is most useful when the sample size is large (\( n \geq 30 \)) because the Central Limit Theorem ensures that the sampling distribution of the sample mean is approximately normal.
- **Simplicity and Speed**: The Z-test is straightforward to compute and requires fewer assumptions about the sample data when compared to other tests, such as the t-test.
- **Standardization**: The Z-test uses standard normal distribution tables, which makes it easy to compare results.

---

### **5. Disadvantages:**

- **Requires Known Population Standard Deviation**: A significant limitation of the Z-test is that it requires knowledge of the population standard deviation, which is often unavailable in practical situations.
- **Not Suitable for Small Sample Sizes**: The Z-test is not ideal when the sample size is small (\( n < 30 \)) because the sampling distribution may not be normal, leading to incorrect conclusions.
- **Assumes Normal Distribution**: The Z-test assumes that the data is normally distributed. If the data is skewed or not normal, the test may not be valid.

---

### **6. Python Implementation:**

Below is a Python implementation of a **One-Sample Z-Test** using `scipy`:


```python
import numpy as np
from scipy.stats import norm

# Given Data
sample_mean = 172
population_mean = 170
population_std = 10
sample_size = 100

# Z-statistic Calculation
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# P-value Calculation for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Z-Statistic: {z_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Decision Rule
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")
```


### **7. Types of Z-Tests in Detail:**

1. **One-Sample Z-Test**:
   - Used to compare the mean of a single sample to a known population mean. This test is valid only when the population standard deviation is known and the sample size is sufficiently large.

2. **Two-Sample Z-Test**:
   - Used to compare the means of two independent samples. It is appropriate when the sample sizes are large and the population standard deviations are known. This test checks if there is a significant difference between the means of the two samples.

3. **Z-Test for Proportions**:
   - Used to compare a sample proportion to a population proportion. This test is used in scenarios where we are interested in proportions (e.g., success rates in experiments, survey responses).

---

### **Factors for Choosing a Test:**

- **Type of Data**:  
  - Z-test is suitable when you have interval/ratio data and the sample size is large or population variance is known.
  
- **Number of Groups**:  
  - One-sample Z-test is for a single group. Two-sample Z-test is used for comparing two independent groups.

- **Sample Size**:  
  - Z-tests are generally appropriate for large sample sizes (n > 30) since the Central Limit Theorem ensures the sampling distribution is normal.

- **Distribution**:  
  - The Z-test assumes that the underlying distribution is normal, or the sample size is large enough to invoke the Central Limit Theorem.

- **Relationship Type**:  
  - Z-tests are typically used for comparing means (or proportions) across groups or comparing sample statistics to population values.

# **F-Test**

The F-test is a statistical test used to compare the variances of two or more groups to determine if they are significantly different from each other. It is commonly used in the context of ANOVA (Analysis of Variance) and regression analysis. The F-test tests whether the group variances are equal or whether there is a significant difference between them.

## **1. Description:**

The F-test uses the ratio of variances to assess whether two populations (or more in the case of ANOVA) have significantly different variances. The test statistic is an F-ratio, which is the ratio of two variances. In ANOVA, the F-test determines if the variability between group means is larger than the variability within the groups.

### **4. Advantages in Detail of All Types**:

#### **One-Way ANOVA F-Test**:
- **Simplicity**: It's straightforward and widely used for comparing multiple groups.
- **Statistical Power**: Tests the null hypothesis that all group means are equal and provides strong evidence when differences exist.

#### **Two-Way ANOVA F-Test**:
- **Interactivity**: It can evaluate not only the main effects of each factor but also the interaction between the factors.
- **Efficiency**: Analyzes multiple factors simultaneously, saving time and effort compared to running separate one-way ANOVA tests.

#### **F-Test for Comparing Two Variances**:
- **Direct Variance Comparison**: It directly tests if the variances of two groups are significantly different.
- **Assumption Testing**: Useful in validating assumptions of other tests, such as homogeneity of variances in ANOVA.

---

### **5. Disadvantages in Detail of All Types**:

#### **One-Way ANOVA F-Test**:
- **Assumptions**: Assumes normality and equal variances. If these assumptions are violated, the results may not be valid.
- **No Information on Which Groups Differ**: ANOVA only tells you if there is a significant difference, but doesn't specify which groups are different from each other.

#### **Two-Way ANOVA F-Test**:
- **Complexity**: Requires careful interpretation, especially when analyzing interaction effects. If the interaction is significant, the main effects may not be meaningful.
- **Assumptions**: Assumes normality and equal variances, similar to one-way ANOVA.

#### **F-Test for Comparing Two Variances**:
- **Sensitive to Sample Size**: The F-test can be sensitive to sample size, especially with unequal sample sizes, which can affect the accuracy of the results.
- **Only Tests for Differences in Variance**: It only tests for variance differences and does not provide any information about the group means.

---

### **6. Python Implementation of All Types**:



```python
   import numpy as np
from scipy import stats
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Example 1: F-Test for Comparing Two Variances
sample1 = np.array([12, 15, 14, 16, 18])
sample2 = np.array([20, 22, 23, 21, 19])

# Compute sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Calculate F-statistic
f_stat = var1 / var2
print(f"F-Statistic for Two Variances: {f_stat}\n")

# Example 2: One-Way ANOVA
group1 = np.array([85, 88, 90, 91, 87])
group2 = np.array([75, 78, 72, 80, 79])
group3 = np.array([65, 68, 70, 72, 74])

# Perform One-Way ANOVA
f_stat_oneway, p_value_oneway = stats.f_oneway(group1, group2, group3)
print(f"One-Way ANOVA F-Statistic: {f_stat_oneway}, P-Value: {p_value_oneway}\n")

# Example 3: Two-Way ANOVA
data = {
    'Scores': [80, 85, 90, 85, 90, 95, 75, 80, 85, 78, 83, 88],
    'Method': ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B'],
    'Classroom': ['1', '1', '1', '2', '2', '2', '1', '1', '1', '2', '2', '2']
}
df = pd.DataFrame(data)

# Fit model for Two-Way ANOVA
model = ols('Scores ~ C(Method) + C(Classroom) + C(Method):C(Classroom)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(f"Two-Way ANOVA Results:\n{anova_table}\n")
```

# ANOVA (Analysis of Variance)

**1. Description:**

ANOVA (Analysis of Variance) is a statistical technique used to compare the means of three or more independent groups to determine if there is a significant difference between them. It works by analyzing the variance within each group and between groups. If the variance between groups is significantly larger than the variance within groups, this suggests that at least one group mean differs from the others.

**3. Numerical Example for Each Type**:

### **One-Way ANOVA Example:**

Consider three groups of students who took different teaching methods:
- **Group 1**: Scores = [60, 65, 70, 75, 80]
- **Group 2**: Scores = [55, 60, 65, 70, 75]
- **Group 3**: Scores = [50, 55, 60, 65, 70]

Calculate the **F-statistic** to check if there's a significant difference in the group means.

### **Two-Way ANOVA Example:**

Consider a study that evaluates the effectiveness of two teaching methods (Factor A) in two different classrooms (Factor B). The data would include scores for each combination of teaching method and classroom.

- **Method A - Classroom 1**: [70, 75, 80]
- **Method A - Classroom 2**: [72, 78, 83]
- **Method B - Classroom 1**: [65, 70, 75]
- **Method B - Classroom 2**: [68, 74, 79]

We calculate **F-statistics** for both factors and their interaction.

### **Repeated Measures ANOVA Example:**

Consider a clinical trial where the blood pressure of 5 patients is measured three times: before, during, and after treatment.

- **Patient 1**: Before = 120, During = 110, After = 115
- **Patient 2**: Before = 130, During = 120, After = 125
- **Patient 3**: Before = 140, During = 130, After = 135
- **Patient 4**: Before = 125, During = 115, After = 120
- **Patient 5**: Before = 135, During = 125, After = 130

We calculate **F-statistics** to check if the treatment has a significant effect.

---

**4. Advantages in Detail:**

### **One-Way ANOVA**:
- **Simplicity**: It's easy to understand and apply when comparing more than two groups.
- **Statistical Power**: Provides a statistical test to compare multiple group means simultaneously, controlling for Type I error.

### **Two-Way ANOVA**:
- **Interactivity**: Helps to analyze the interaction effect between two factors.
- **Efficiency**: It allows for analyzing the effects of two variables at once, saving time and resources compared to running two separate one-way ANOVA tests.

### **Repeated Measures ANOVA**:
- **Same Subjects**: Because the same subjects are used across different treatments or time points, variability due to individual differences is controlled.
- **Power**: Typically more powerful than a one-way ANOVA with separate groups, as it accounts for within-subject correlations.

---

**5. Disadvantages in Detail:**

### **One-Way ANOVA**:
- **Assumptions**: Assumes homogeneity of variances across groups. If the variances are unequal, results may not be reliable.
- **No Interaction Effects**: Only one factor is considered, so interactions between multiple factors cannot be evaluated.

### **Two-Way ANOVA**:
- **Complexity**: The model is more complex and requires more careful interpretation, especially in the presence of interactions between factors.
- **Assumptions**: Requires normality and equal variances within groups for each factor combination.

### **Repeated Measures ANOVA**:
- **Sphericity Assumption**: Assumes that the variances of the differences between all combinations of related groups are equal, which may not always hold.
- **Complexity**: Interpreting the results can be more difficult due to the correlation of measurements within subjects.

---

**6. Python Implementation of All Types**:



```python
import numpy as np
from scipy import stats

# One-Way ANOVA Example
group1 = np.array([60, 65, 70, 75, 80])
group2 = np.array([55, 60, 65, 70, 75])
group3 = np.array([50, 55, 60, 65, 70])

# Perform one-way ANOVA
f_stat, p_value = stats.f_oneway(group1, group2, group3)
print(f"One-Way ANOVA:\nF-Statistic: {f_stat}\nP-Value: {p_value}\n")

# Two-Way ANOVA Example (using statsmodels)
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Example Data: Teaching Methods (A/B) and Classrooms (1/2)
data = {
    'Scores': [70, 75, 80, 72, 78, 83, 65, 70, 75, 68, 74, 79],
    'Method': ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B'],
    'Classroom': ['1', '1', '1', '2', '2', '2', '1', '1', '1', '2', '2', '2']
}

df = pd.DataFrame(data)

# Fit model
model = ols('Scores ~ C(Method) + C(Classroom) + C(Method):C(Classroom)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(f"Two-Way ANOVA:\n{anova_table}\n")

# Repeated Measures ANOVA (using statsmodels)
# Data structure and method for repeated measures analysis can be complex, typically involving mixed-effects models.
```
