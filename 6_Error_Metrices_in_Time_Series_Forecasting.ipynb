# **Mean Absolute Error (MAE)** in Time Series Forecasting


#### **Description:**
Mean Absolute Error (MAE) is a metric that quantifies the average absolute difference between the actual observed values and the predicted values in a dataset. It measures the magnitude of errors in a set of predictions, providing a straightforward assessment of forecast accuracy.

Key Features of MAE:
- **Scale Sensitivity:** The error is expressed in the same units as the data, making it easy to interpret.
- **Error Focus:** MAE disregards the direction of errors (whether the prediction is higher or lower than the actual value) by taking the absolute value of each error.
- **Use Case Suitability:** MAE is ideal when all errors, regardless of their size, should be treated equally, making it less sensitive to outliers compared to other metrics like Mean Squared Error (MSE).

---

#### **When to Use:**
- **Simplicity:** MAE is useful when simplicity and interpretability are priorities.
- **Balanced Error Measurement:** It is suitable for cases where both small and large errors are equally important.
- **Real-World Applications:** Ideal for fields like demand forecasting, energy consumption, and temperature predictions.

---

#### **Interpretation of Results:**
- A **lower MAE** indicates better predictive accuracy.
- The metric provides a direct measure of how much predictions deviate from actual values, in the same units as the data.

---

#### **Advantages:**
1. **Intuitive and Easy to Interpret:** Results are in the same unit as the data, making MAE straightforward to understand.
2. **Equal Error Weighting:** MAE does not disproportionately penalize larger errors.
3. **Robust Across Applications:** It is versatile and widely applicable across industries.

---

#### **Disadvantages:**
1. **Non-Differentiable:** The metric cannot be optimized directly in gradient-based methods due to its non-differentiable nature at zero.
2. **Limited Error Information:** MAE does not indicate the direction of errors or highlight the distribution of residuals.
3. **Uniform Treatment of Errors:** It may not be ideal in cases where larger errors should be penalized more heavily.

---

#### **Python Code:**

Here’s how to calculate MAE in Python:

```python
import numpy as np
from sklearn.metrics import mean_absolute_error

# Actual and predicted values
actual_values = np.array([100, 120, 130, 140])
predicted_values = np.array([105, 115, 125, 150])

# Calculate MAE
mae = mean_absolute_error(actual_values, predicted_values)
print(f"Mean Absolute Error (MAE): {mae}")
```

---

#### **Applications in Time Series Forecasting:**
- **Comparing Models:** Assessing the accuracy of different forecasting models, such as ARIMA, SARIMA, or LSTM.
- **Evaluating Business Impact:** Determining how forecast errors affect real-world decision-making.
- **Optimizing Predictions:** Using MAE as a benchmark during hyperparameter tuning and model selection.

# **Mean Squared Error (MSE)**


#### **Description**
Mean Squared Error (MSE) is a common metric used to measure the average squared difference between actual and predicted values in forecasting or regression models. It gives higher weight to larger errors due to squaring, which makes it sensitive to outliers. MSE is widely used in time series forecasting to assess the accuracy of models.

---

#### **Mathematical Formula**

image.png


---

#### **Numerical Example**
| Time Period | Actual Value (\(y_i\)) | Predicted Value (\(\hat{y}_i\)) | Squared Error (\((y_i - \hat{y}_i)^2\)) |
|-------------|-------------------------|----------------------------------|-----------------------------------------|
| 1           | 100                     | 105                              | 25                                      |
| 2           | 120                     | 115                              | 25                                      |
| 3           | 130                     | 125                              | 25                                      |
| 4           | 140                     | 150                              | 100                                     |

image.png
---

---

#### **When to Use**
- When large errors need to be penalized more heavily than smaller ones.
- When outlier detection and impact are important.
- In machine learning models, to optimize loss functions like MSE.

---

#### **Python Code**

```python
import numpy as np

# Actual and predicted values
actual_values = np.array([100, 120, 130, 140])
predicted_values = np.array([105, 115, 125, 150])

# Calculate MSE
mse = np.mean((actual_values - predicted_values) ** 2)
print(f"Mean Squared Error (MSE): {mse}")
```


---

#### **Application in Time Series**
- Evaluates the predictive accuracy of forecasting models (e.g., ARIMA, LSTM, Prophet).
- Used in model selection and hyperparameter tuning during training.

---

#### **Interpretation of Results**
- A lower MSE indicates a better fit of the model to the data.
- Sensitive to the scale of data; cannot compare datasets with different units directly.
- Higher values indicate large prediction errors.

---

#### **Advantages**
1. Penalizes large errors heavily, making it effective in detecting poorly performing models.
2. Easy to compute and widely supported in statistical software.
3. Useful in gradient-based optimization due to smoothness.

---

#### **Disadvantages**
1. Sensitive to outliers because errors are squared, disproportionately affecting results.
2. Not scale-invariant; results depend on the unit of measurement.
3. Harder to interpret compared to MAE since the errors are squared.

# **Mean Absolute Scaled Error (MASE) with regards to Time Series Forecasting**

#### **Description**
The **Mean Absolute Scaled Error (MASE)** is an error metric used for evaluating the accuracy of time series forecasts. It is designed to address some limitations of traditional error metrics like **Mean Absolute Percentage Error (MAPE)** and **Mean Squared Error (MSE)**. MASE is scaled, meaning it normalizes the forecast errors by comparing them with a baseline forecast. This baseline forecast is typically a "naive" forecast, where the predicted value for each time point is equal to the value of the previous time point.

Unlike traditional metrics like MAPE, MASE is **scale-independent**, making it more suitable for comparing forecasts across different datasets or when the time series data has varying scales.

The primary advantage of MASE is that it handles zero and near-zero values in the time series more effectively than other error metrics. MASE also does not suffer from the problem of undefined errors, which is an issue in MAPE when actual values are zero.

#### **Numerical Example**

Let’s consider a time series with actual and predicted values:

| Time Period | Actual Value | Predicted Value  | Absolute Error  |
|-------------|-------------------------|----------------------------------|------------------------------------|

| 2           | 120                     | 115                              | 5                                  |
| 3           | 130                     | 125                              | 5                                  |
| 4           | 140                     | 150                              | 10                                 |

**Step 1**: Calculate the mean absolute error of the naive forecast. The naive forecast is the actual value from the previous time point:

| Time Period | Actual Value (\(y_i\)) | Naive Forecast (\(y_{i-1}\)) | Absolute Error of Naive Forecast |
|-------------|-------------------------|-----------------------------|---------------------------------|
| 1           | 100                     | -                           | -                               |
| 2           | 120                     | 100                         | 20                              |
| 3           | 130                     | 120                         | 10                              |
| 4           | 140                     | 130                         | 10                              |

---

#### **When to Use**
- **MASE** is most useful when comparing different forecasting models or when dealing with datasets that have varying scales. It can be used to evaluate forecasting performance on both short-term and long-term forecasts.
- MASE is also preferred when the series contains zero or near-zero values, as it avoids the problems of undefined errors like MAPE.
- It is often used in **time series forecasting** for applications such as demand forecasting, sales prediction, and inventory management.

---

#### **Python Code**

```python
import numpy as np

# Actual and predicted values
actual_values = np.array([100, 120, 130, 140])
predicted_values = np.array([105, 115, 125, 150])

# Calculate the absolute error for each prediction
absolute_errors = np.abs(actual_values - predicted_values)

# Calculate the mean absolute error of the naive forecast
naive_forecast_errors = np.abs(np.diff(actual_values))
naive_mae = np.mean(naive_forecast_errors)

# Calculate MASE
mase = np.mean(absolute_errors) / naive_mae
print(f"Mean Absolute Scaled Error (MASE): {mase:.4f}")
```


---

#### **Application in Time Series**

- **MASE** is especially useful for **model comparison** in time series forecasting, as it allows comparing models on different datasets with different scales.
- It is often used in **demand forecasting**, **sales prediction**, and **energy consumption** forecasting, where comparing models across different time series datasets is essential.

---

#### **Interpretation of Results**
- **Lower MASE**: A lower MASE indicates better model accuracy, as the predicted values are closer to the actual values relative to the baseline (naive forecast).
- **Higher MASE**: A higher MASE suggests that the forecast model performs worse than the naive forecast, indicating poor model performance.

---

#### **Advantages**
1. **Scale-Independent**: MASE is not sensitive to the scale of the data, making it suitable for comparing models across datasets with different units and magnitudes.
2. **Handles Zero Values**: Unlike MAPE, MASE handles time series with zero or near-zero values well, as it compares errors relative to a baseline model.
3. **Robustness**: It is robust against cases where the model is a poor fit for the data, providing a normalized error value.

---

#### **Disadvantages**
1. **Dependence on Naive Model**: MASE depends on the naive model for scaling, so its performance can be misleading if the naive model itself is not a good fit for the data.
2. **Less Intuitive**: For some users, MASE might not be as intuitive to interpret as other metrics like RMSE or MAPE.
3. **Can Be Sensitive in Certain Scenarios**: If the series has small values but the naive model does not perform well, MASE might give a misleading impression of model performance.

# **Root Mean Squared Error (RMSE)**

#### **Description**  
Root Mean Squared Error (RMSE) is the square root of the Mean Squared Error (MSE). It measures the average magnitude of prediction errors, giving higher weight to larger errors due to the squaring operation in MSE. RMSE is expressed in the same unit as the target variable, making it easier to interpret in real-world terms.

---

#### **Mathematical Formula**

image.png


#### **Numerical Example**

| Time Period | Actual Value (\(y_i\)) | Predicted Value (\(\hat{y}_i\)) | Squared Error (\((y_i - \hat{y}_i)^2\)) |
|-------------|-------------------------|----------------------------------|-----------------------------------------|
| 1           | 100                     | 105                              | 25                                      |
| 2           | 120                     | 115                              | 25                                      |
| 3           | 130                     | 125                              | 25                                      |
| 4           | 140                     | 150                              | 100                                     |

image.png

---

#### **When to Use**
- When you want to evaluate the average magnitude of prediction errors in the same unit as the original data.
- When large errors need to be penalized more heavily than smaller ones.
- When comparing the performance of different models with similar scales.

---

#### **Python Code**

```python
import numpy as np

# Actual and predicted values
actual_values = np.array([100, 120, 130, 140])
predicted_values = np.array([105, 115, 125, 150])

# Calculate RMSE
rmse = np.sqrt(np.mean((actual_values - predicted_values) ** 2))
print(f"Root Mean Squared Error (RMSE): {rmse}")
```


---

#### **Application in Time Series**
- Widely used to assess forecasting model performance (e.g., ARIMA, ETS, LSTM).
- Helps compare models on the same dataset or across datasets with similar scales.
- Useful for identifying models that minimize large forecast errors.

---

#### **Interpretation of Results**
- **Lower RMSE**: Indicates better model performance and closer predictions to actual values.
- **Higher RMSE**: Signals larger deviations between predictions and actual values.
- RMSE is directly tied to the scale of the data and cannot be compared across datasets with different units.

---

#### **Advantages**
1. **Easy to Interpret**: Expressed in the same units as the target variable.  
2. **Sensitive to Large Errors**: Effective for models where large deviations are critical.  
3. **Widely Used**: A standard metric in time series and regression tasks.  

---

#### **Disadvantages**
1. **Sensitive to Outliers**: Squaring the errors amplifies the impact of outliers.  
2. **Not Scale-Invariant**: Depends on the unit of measurement, making cross-dataset comparisons difficult.  
3. **Does Not Reflect Bias**: Fails to differentiate between over- and under-predictions.

# **Normalized Root Mean Squared Error (NRMSE)**

#### **Description**
Normalized Root Mean Squared Error (NRMSE) is a variation of RMSE (Root Mean Squared Error) that scales the error based on the range or the mean of the actual values. This normalization makes NRMSE a unit-free metric, enabling comparisons across datasets with different scales. NRMSE provides a clearer understanding of the error magnitude in relation to the data, and it is useful when comparing models with different output scales or when you want to assess the error in a standardized way.

---

#### **Mathematical Formula**
The formula for NRMSE is:

image.png

---

#### **Numerical Example**

| Time Period | Actual Value (\(y_i\)) | Predicted Value (\(\hat{y}_i\)) | Squared Error \((y_i - \hat{y}_i)^2\) |
|-------------|-------------------------|----------------------------------|------------------------------------|
| 1           | 100                     | 105                              | 25                                 |
| 2           | 120                     | 115                              | 25                                 |
| 3           | 130                     | 125                              | 25                                 |
| 4           | 140                     | 150                              | 100                                |


image.png

---

#### **When to Use**
- When you want a unit-free metric that allows comparisons across different datasets.
- Particularly useful when the data has a wide range of values.
- Suitable for comparing models across different time series with varying units or scales.

---

#### **Python Code**

```python
import numpy as np

# Actual and predicted values
actual_values = np.array([100, 120, 130, 140])
predicted_values = np.array([105, 115, 125, 150])

# Calculate RMSE
rmse = np.sqrt(np.mean((actual_values - predicted_values)**2))

# Calculate Range
range_actual = np.max(actual_values) - np.min(actual_values)

# Calculate NRMSE
nrmse = rmse / range_actual
print(f"Normalized Root Mean Squared Error (NRMSE): {nrmse:.4f}")
```


---

#### **Application in Time Series**
- NRMSE is often used in **time series forecasting** to assess the model’s prediction accuracy.
- It is particularly useful when comparing **forecasting models** on datasets that vary significantly in scale.
- Helps compare models for tasks like **sales prediction**, **energy demand forecasting**, or **weather prediction**, where the magnitude of data values can differ.

---

#### **Interpretation of Results**
- **Lower NRMSE**: Indicates better model performance, meaning the predicted values are close to the actual values relative to the range of the data.
- **Higher NRMSE**: Indicates poor model performance, meaning the model is not predicting accurately in relation to the scale of the data.

---

#### **Advantages**
1. **Unit-Free**: Allows comparison between datasets with different units or scales.
2. **Easy to Interpret**: Gives a clear measure of prediction accuracy relative to the data’s range.
3. **Suitable for Large Variations**: Works well when the data has a wide range of values.

---

#### **Disadvantages**
1. **Sensitive to Range**: NRMSE can be misleading when the range is very small, especially if the actual values are close to each other.
2. **Assumes Normality**: Like RMSE, NRMSE assumes a normal distribution of errors, which might not always be true.
3. **Outliers Impact**: Just like RMSE, outliers can have a disproportionate impact on NRMSE.

# **Mean Absolute Percentage Error (MAPE)**

#### **Description**  
Mean Absolute Percentage Error (MAPE) is a measure of forecasting accuracy that expresses errors as a percentage of the actual values. It is calculated as the average of the absolute percentage errors between the predicted and actual values. MAPE is widely used because it provides a simple, unit-free metric to evaluate model performance. However, it is sensitive to the scale of the actual values, and division by small actual values can distort results.

---

#### **Mathematical Formula**

image.png






| Time Period | Actual Value | Predicted Value | Absolute Percentage Error |
|-------------|-------------------------|----------------------------------|-------------------------------------------------------------------------------------|
| 1           | 100                     | 105                              | \( (100 - 105)/100 x 100 = 5.0\% \)                       |
| 2           | 120                     | 115                              | (120 - 115)/120 x 100 = 4.17\% \)                     |
| 3           | 130                     | 125                              | (130 - 125)/130 x100 = 3.85\% \)                     |
| 4           | 140                     | 150                              | (140 - 150)/140 x 100 = 7.14\% \)


image.png

---

#### **When to Use**
- When you need a unit-free metric for model comparison.
- For evaluating models where prediction accuracy is proportional to the actual values.
- Particularly useful when the target variable spans multiple magnitudes.

---

#### **Python Code**

```python
import numpy as np

# Actual and predicted values
actual_values = np.array([100, 120, 130, 140])
predicted_values = np.array([105, 115, 125, 150])

# Calculate MAPE
mape = np.mean(np.abs((actual_values - predicted_values) / actual_values)) * 100
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
```

---

#### **Application in Time Series**
- Widely used in business forecasting, such as demand and sales prediction.
- Helps evaluate models for datasets with varying magnitudes of values.
- Particularly useful in applications where percentage accuracy is crucial.

---

#### **Interpretation of Results**
- **Lower MAPE**: Indicates better model performance, with predictions closely matching the actual values.  
- **Higher MAPE**: Indicates poorer model performance.  

---

#### **Advantages**
1. **Easy to Understand**: Expresses errors as percentages, making results interpretable for non-technical stakeholders.  
2. **Unit-Free**: Allows comparison across datasets with different scales.  
3. **Widely Used**: Standard metric in time series forecasting tasks.  

---

#### **Disadvantages**
1. **Division by Zero Issue**: Small actual values can lead to distorted or undefined results.  
2. **Sensitivity to Scale**: Heavily influenced by the magnitude of actual values.  
3. **Skewness**: Penalizes overestimation and underestimation differently, leading to biases in evaluation.

# **Symmetric Mean Absolute Percentage Error (sMAPE) with regards to Time Series Forecasting**

#### **Description**
Symmetric Mean Absolute Percentage Error (sMAPE) is an accuracy metric used in time series forecasting that measures the accuracy of a forecast. It is an improvement over the Mean Absolute Percentage Error (MAPE) by addressing some of its shortcomings, such as handling cases when the actual value is close to zero. The key feature of sMAPE is that it uses a symmetric formula, meaning it treats both over-predictions and under-predictions equally, making it more balanced and robust compared to MAPE.

Unlike MAPE, which can become undefined when the actual value is zero, sMAPE avoids this issue by taking the absolute difference between the actual and predicted values and normalizing it by the sum of actual and predicted values.

---

#### **Mathematical Formula**

The formula for **sMAPE** is:

image.png



---

#### **Numerical Example**

Let’s consider a time series data with actual and predicted values:

| Time Period | Actual Value  | Predicted Value  | Absolute Error | Normalization |
|-------------|-------------------------|----------------------------------|------------------------------------|-------------------------------------------------|
| 1           | 100                     | 105                              | 5                                  | 102.5                                           |
| 2           | 120                     | 115                              | 5                                  | 117.5                                           |
| 3           | 130                     | 125                              | 5                                  | 127.5                                           |
| 4           | 140                     | 150                              | 10                                 | 145                                             |

image.png


---

#### **When to Use**
- **sMAPE** is useful when comparing forecasting models, especially when the data has values close to zero or contains very large values.
- It is often used in **time series forecasting** tasks, such as predicting demand, sales, or stock prices, because it accounts for both overestimation and underestimation symmetrically.
- It is best used when the magnitude of the forecasted and actual values are similar across different data points.

---

#### **Python Code**

```python
import numpy as np

# Actual and predicted values
actual_values = np.array([100, 120, 130, 140])
predicted_values = np.array([105, 115, 125, 150])

# Calculate sMAPE
sMAPE = np.mean(2 * np.abs(actual_values - predicted_values) / (np.abs(actual_values) + np.abs(predicted_values))) * 100
print(f"Symmetric Mean Absolute Percentage Error (sMAPE): {sMAPE:.2f}%")
```


---

#### **Application in Time Series**
- **sMAPE** is widely used in time series forecasting, especially for tasks where prediction errors need to be analyzed symmetrically (i.e., both over- and under-predictions should be treated equally).
- It is commonly used in business and economic forecasting, such as predicting sales, stock prices, energy consumption, and demand forecasting in retail, energy, and manufacturing sectors.

---

#### **Interpretation of Results**
- **Lower sMAPE**: A lower sMAPE value indicates better forecasting accuracy. It means the predicted values are closer to the actual values relative to their average magnitude.
- **Higher sMAPE**: A higher sMAPE suggests poor accuracy, indicating that the model's predictions deviate significantly from the actual values.

---

#### **Advantages**
1. **Symmetry**: sMAPE treats both over- and under-predictions equally, unlike MAPE, which can be biased toward large actual values.
2. **Unit-Free**: Like other percentage-based errors, sMAPE is a unit-free metric, allowing comparisons across datasets with different units.
3. **More Robust**: Handles small values of actuals more effectively than MAPE since it is symmetric.

---

#### **Disadvantages**
1. **Zero-Handling**: While it handles zero values better than MAPE, the formula still uses the sum of actual and predicted values in the denominator, which can be problematic when both actual and predicted values are near zero.
2. **Less Intuitive for Some**: sMAPE can be harder to interpret compared to metrics like RMSE, especially for users who are more familiar with error metrics in their raw scale.
3. **Can Be Biased in Certain Scenarios**: If the actual and predicted values are small but not close to zero, sMAPE might still behave inconsistently.

# **R² (Coefficient of Determination) and Adjusted R² for Time Series Forecasting**

#### **1. R² (Coefficient of Determination)**

##### **Description**
The **R²** value, or **Coefficient of Determination**, is a key statistical measure that indicates the proportion of the variance in the dependent variable that is predictable from the independent variables in a regression model. It is widely used to assess the goodness of fit of the model. For time series forecasting, R² helps to measure how well the model's predicted values align with the actual time series data.

In time series forecasting, R² is used to evaluate how much of the variance in future values is explained by the past data points that are used as predictors in the model (e.g., in ARIMA or regression models).

- **R² = 1**: The model explains 100% of the variance in the time series data.
- **R² = 0**: The model explains none of the variance, i.e., the model is no better than simply predicting the mean of the dependent variable.

##### **Mathematical Formula**
The formula for R² is:

##### **Numerical Example**
Assume we have the following data:

| Time Period | Actual Value (\(y_t\)) | Predicted Value (\(\hat{y}_t\)) |
|-------------|-------------------------|--------------------------------|
| 1           | 100                     | 105                            |
| 2           | 120                     | 115                            |
| 3           | 130                     | 125                            |

1. **Step 1**: Calculate the squared differences between actual values and predicted values.
2. **Step 2**: Compute the squared differences between actual values and the mean of the actual values.
3. **Step 3**: Apply the formula.

**Python Code Example for R²**:

```python
import numpy as np
from sklearn.metrics import r2_score

# Actual and predicted values
y_true = np.array([100, 120, 130, 140])
y_pred = np.array([105, 115, 125, 138])

# Calculate R² using sklearn
r2 = r2_score(y_true, y_pred)
print(f"R²: {r2:.4f}")
```

#### **2. Adjusted R² (Adjusted Coefficient of Determination)**

##### **Description**
The **Adjusted R²** is a modified version of R² that adjusts for the number of explanatory variables (predictors) in the model. It is used to determine how well the model fits the data, while accounting for the number of predictors. Unlike R², which can be artificially inflated when more variables are added to the model, the Adjusted R² provides a more reliable measure of model performance when dealing with multiple predictors.

In time series forecasting, Adjusted R² is useful for evaluating models with multiple predictors (like in multivariate time series models). It prevents overfitting by penalizing the inclusion of unnecessary predictors that do not improve the model significantly.

# **GMRAE (Geometric Mean Relative Absolute Error)**

#### **Description:**
The **GMRAE** is a metric used to measure the accuracy of a forecasting model, especially when there are outliers or large deviations between the predicted and actual values. It is a modification of the traditional **Mean Absolute Error (MAE)**, which reduces the impact of extreme errors by using the geometric mean instead of the arithmetic mean. The GMRAE is often used when working with datasets where large errors can skew the performance of other accuracy metrics like MAE or RMSE.

In GMRAE, the error is computed by taking the relative absolute errors for each point, followed by computing the geometric mean of these relative errors.

#### **Numerical Example:**

Assume the following time series data:

| Time Period | Actual Value (\(y_i\)) | Predicted Value (\(\hat{y}_i\)) |
|-------------|------------------------|---------------------------------|
| 1           | 100                    | 110                             |
| 2           | 120                    | 125                             |
| 3           | 130                    | 135                             |
| 4           | 140                    | 145                             |

1. **Step 1**: Calculate the relative absolute errors:

| Time Period | Actual Value (\(y_i\)) | Predicted Value (\(\hat{y}_i\)) | Absolute Error | Relative Absolute Error |
|-------------|------------------------|---------------------------------|----------------|-------------------------|
| 1           | 100                    | 110                             | 10             | 0.1                     |
| 2           | 120                    | 125                             | 5              | 0.0417                  |
| 3           | 130                    | 135                             | 5              | 0.0385                  |
| 4           | 140                    | 145                             | 5              | 0.0357                  |

2. **Step 2**: Take the logarithm of each relative absolute error:

#### **When to Use:**
- **GMRAE** is particularly useful in cases where the data has significant outliers or large variations, as the geometric mean reduces the influence of large errors.
- It is effective when evaluating models that are expected to handle large deviations gracefully.

#### **Interpretation of Results:**
- A **GMRAE value of 1** implies perfect forecasting, as the error is relative to the actual value.
- **GMRAE < 1** indicates that the model performs better than the baseline (predicting the mean value).
- **GMRAE > 1** suggests that the model performs worse than the baseline.

#### **Advantages:**
1. **Handles Outliers**: GMRAE is less sensitive to extreme outliers compared to traditional metrics like MAE and RMSE because it uses the geometric mean.
2. **Comparative Measurement**: Provides a clear indication of whether a model is outperforming a simple baseline model (such as predicting the mean value).
3. **Intuitive**: The interpretation of GMRAE is more intuitive when the values are scaled between 0 and 1.

#### **Disadvantages:**
1. **Sensitive to Zero Values**: If the actual value \( y_i \) is zero, it will lead to undefined values in the formula since division by zero occurs.
2. **Complex Calculation**: Requires more steps to calculate than basic metrics like MAE or MSE, though still computationally manageable.
3. **Not Standard in Literature**: It is not as widely recognized or used as other metrics like MSE or MAE, which could make it harder to compare across different contexts.

#### **Python Code:**

```python
import numpy as np

def gmrae(y_true, y_pred):
    # Calculate the absolute errors and relative absolute errors
    rel_abs_errors = np.abs(y_true - y_pred) / np.abs(y_true)

    # Take the natural logarithm of each relative absolute error
    log_errors = np.log(rel_abs_errors)

    # Compute the geometric mean of the logarithms
    gmrae_value = np.exp(np.mean(log_errors))

    return gmrae_value

# Example data
y_true = np.array([100, 120, 130, 140])
y_pred = np.array([110, 125, 135, 145])

# Compute GMRAE
gmrae_value = gmrae(y_true, y_pred)
print(f"GMRAE: {gmrae_value:.4f}")
```

### **Summary:**
- **MASE (Mean Absolute Scaled Error)** is a good choice when comparing different forecasting models or datasets with different scales, as it is scale-independent and less sensitive to outliers.
- **RMSE (Root Mean Squared Error)** is suitable when large errors need to be penalized more. It is widely used but highly sensitive to outliers, making it less ideal when the data contains extreme values.
- **MAPE (Mean Absolute Percentage Error)** is intuitive and interpretable but can be problematic with small actual values, as it becomes extremely sensitive and may produce misleading results.
- **sMAPE (Symmetric Mean Absolute Percentage Error)** is a better alternative to MAPE when a symmetric error metric is required, especially when handling zero and negative values. However, it can still be affected by large deviations and is more complex to compute.
